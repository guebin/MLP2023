# 4. 적합

```python
df_lst = [df for k,df in df_train.groupby(['country','store','product'])]
df_lst[0]
```

```python
df_lst[0].loc[:,['date','country','num_sold']]
```

```python
# i=0 
tsdf_train = TimeSeriesDataFrame(df_lst[0].loc[:,['date','country','num_sold']], id_column='country', timestamp_column='date')
```

```python
predictr = TimeSeriesPredictor(
    target="num_sold",
    freq='D',
    prediction_length=365
)
```

```python
predictr.fit(tsdf_train,time_limit=60)
```

```python
predictr.leaderboard()
```

```python
tsdf_train.tail()
```

```python
predictr.predict(tsdf_train)['mean'].plot()
```
